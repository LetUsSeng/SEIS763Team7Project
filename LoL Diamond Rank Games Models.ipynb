# Reading in the data

```python
import pandas as pd
lol_games = pd.read_csv("high_diamond_ranked_10min.csv")
lol_games
```

```python
lol_games = lol_games.drop('gameId', axis=1)
lol_games
```

```python
X = lol_games.drop('blueWins', axis=1)
y = lol_games['blueWins'].values
X
```

```python
#Splitting the data into Training Set and Test Set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=0)
X_test
```

# Decision Tree

```python
#Fitting Classifier to Training Set. Create a classifier object here and call it classifierObj
from sklearn.tree import DecisionTreeClassifier
classifierObj = DecisionTreeClassifier(criterion='entropy', random_state=0)
classifierObj.fit(X_train,y_train)
```

```python
#Making predictions on the Test Set
decision_tree_pred = classifierObj.predict(X_test)
```

```python
#Evaluating the predictions using a Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, decision_tree_pred)
cm
```

```python
classifierObj.score(X_test,y_test)
# a single decision tree gets a score of 64.6%
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
classifierObj = RandomForestClassifier(n_estimators=500, criterion='entropy', max_depth=8, min_samples_leaf=20, random_state=0)
classifierObj.fit(X_train,y_train)
```

```python
#Making predictions on the Test Set
random_forest_pred = classifierObj.predict(X_test)
```

```python
classifierObj.score(X_test,y_test)
# 72.3%
```

# KNN

```python
#Normalizing the features
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train_normalized = sc_X.fit_transform(X_train)
X_test_normalized = sc_X.transform(X_test)
```

```python
#Fitting Classifier to Training Set. Create a classifier object here and call it classifierObj
from sklearn.neighbors import KNeighborsClassifier
classifierObj = KNeighborsClassifier(n_neighbors=50, p=2, metric='minkowski')
classifierObj.fit(X_train_normalized, y_train)
```

```python
#Making predictions on the Test Set
knn_pred = classifierObj.predict(X_test_normalized)
```

```python
classifierObj.score(X_test_normalized,y_test)
# 71.5%
```

# Logistic Regression

```python
#Fitting Logistic Regression to Training Set
from sklearn.linear_model import LogisticRegression
classifierObj = LogisticRegression(random_state=0)
classifierObj.fit(X_train, y_train)
```

```python
classifierObj.coef_
```

```python
#Making predictions on the Test Set
logistic_reg_pred = classifierObj.predict(X_test)
```

```python
classifierObj.score(X_test,y_test)
# 72%
```

# Naive Bayes

```python
#Fitting Classifier to Training Set. Create a classifier object here and call it classifierObj
from sklearn.naive_bayes import GaussianNB
classifierObj = GaussianNB()
classifierObj.fit(X_train, y_train)
```

```python
#Making predictions on the Test Set
naive_bayes_pred = classifierObj.predict(X_test)
```

```python
classifierObj.score(X_test,y_test)
# 72.1%
```

# SVM

```python
#Fitting Classifier to Training Set. Create a classifier object here and call it classifierObj
# can use linear, poly, rbf, sigmoid,or precomputed kernels
from sklearn.svm import SVC
classifierObj = SVC(C=2, kernel='linear')
classifierObj.fit(X_train, y_train)
```

```python
#Making predictions on the Test Set
svm_pred = classifierObj.predict(X_test)
```
